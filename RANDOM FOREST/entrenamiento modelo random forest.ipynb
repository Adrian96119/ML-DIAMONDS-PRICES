## ENTRENAMIENTO Y PREDICCION RANDOM FOREST 

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
```

```python
train = pd.read_csv("./train_para_entrenar.csv")
test = pd.read_csv("../predict.csv")
```

```python

train
```

```python
#entrenemos el modelo con 1000 arboles de decision
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
```

```python
modelo = rf.fit(train.drop("price",axis=1),train["price"])
```

```python
#tenemos que ajustar nuestro data de test para poder hacer predicciones.
testt = test.drop(["id","x","y","z"],axis=1)
```

```python
#convertimos a ordinales las columnas categoricas del test.
test_dummies = pd.get_dummies(testt)
```

```python
#hacemos las predicciones
```

```python
y_pred = modelo.predict(test_dummies)
```

```python
fin_prediccion = pd.DataFrame({"id":test["id"],"price":y_pred})
```

```python
fin_prediccion.to_csv("./prediccion_final.csv",index=False) #exportamos el modelo.
```
