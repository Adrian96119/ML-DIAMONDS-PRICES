```python
from sklearn.linear_model import LinearRegression

import pandas as pd
```

```python
test = pd.read_csv("../predict.csv")
```

```python
train = pd.read_csv("./modelo_regresionLineal.csv")
```

```python
lr = LinearRegression()
```

```python
modelo = lr.fit(train.drop("price",axis=1),train["price"])
```

```python
X_test = test.drop(["id","cut","color","clarity","x","y","z"],axis=1) #nivelo el X de prueba para que mi modelo haga predicciones.
```

```python
y_pred =modelo.predict(X_test)
```

```python

fin_prediccion = pd.DataFrame({"id":test["id"],"price":y_pred})
```

```python
fin_prediccion.to_csv (".\TEST1.csv",index = False) #exporto el csv con la prediccion y los ids.
```
