```python
import pandas as pd
```

```python
train = pd.read_csv("./train.csv")
```

```python
#renombramos las columnas
train.rename(columns={'carat': 'quilates','cut':'calidad corte','depth':'porcentaje profundidad','table': 'ancho superior','x':'longitud mm','y':'ancho mm','z':'profundidad mm'},inplace=True)
```

```python
#comprobamos que no hay valores nulos.
train.isnull().sum()
```

```python
#comprobamos que no hay valores de cero
train.describe() #observamos que efectiamente hay ceros en longitud, ancho y profundidad.
```

```python
#vamos a sustituir los ceros por valores predichos. y para ello vamos a importar unas cuantas librerias.
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

```

**prediccion profundidad**

```python
lr_profundidad = LinearRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(train[["quilates","ancho superior"]],train["profundidad mm"],train_size =0.8,test_size = 0.2)
```

```python
modelo_profundidad = lr_profundidad.fit(X_train,y_train)
y_pred_profts = modelo_profundidad.predict(X_test) #predecimos la prueba
y_pred_proftr =modelo_profundidad.predict(X_train) #predecimos el entrenamiento
```

```python
profundidad_predicha_ts = pd.DataFrame({"real":y_test,"pred":y_pred_profts})
profundidad_predicha_tr = pd.DataFrame({"real":y_train,"pred":y_pred_proftr})
```
